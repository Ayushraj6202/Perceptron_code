```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=1000, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
data=pd.DataFrame(X,y,columns=['a','b'])
data.head()
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
```

```python
def step(a):
    if (a>0):
        return 1
    return 0
print(X.shape[0])
```

```python
def Preceptron(X,y):
    lr=0.1
    X=np.insert(X,0,1,axis=1)
    weight=np.ones(3);
    for i in range(1000):
        ind = np.random.randint(0,X.shape[0])
        y_=step(np.dot(X[ind],weight))
        #chaning line slope or intercept by a little 1000 times to reach best line
        weight+=(lr*(y[ind]-y_)*X[ind])
    return weight[0],weight[1:]
```

```python
intercept_,coef_ = Preceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,5))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
